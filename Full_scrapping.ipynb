### Scrapping items on different pages

#### This code take price and description from a search in a store website and take all elements I want(in this cases I want 15 page items elements, create a .csv and make a dataframe

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import pandas as pd

#################

filename='scrap.csv'
headers='description,price\n'
f=open(filename,'w')
f.write(headers)
    
for i in range(1,15):

        page='https://www.aliexpress.com/wholesale?site=glo&g=y&SearchText=laptop+charger&page={}&CatId=14191109&initiative_id=SB_20171020081119&needQuery=n'.format(i)
        uClient= uReq(page)
        page_html= uClient.read()
        uClient.close()
        
        page_soup=soup(page_html,'html.parser')
        articulos= page_soup.find_all('div',{'class':'info'})  
        
        
        
    
        
        for articulo in articulos:
        
            info=articulo.find('a',{'class':'history-item product '})
            try:
                info=info['title']

            except:
                continue
        
            precio=articulo.find('span',{'class':'value'})
            precio=precio.text
            info=info.replace(',','')
            precio=precio.replace('US $','')

            f.write(info + ',' + precio + '\n')

            

f.close()

df=pd.read_csv(filename)
```
